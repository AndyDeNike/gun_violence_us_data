# Gun violence in the USA 2013 - 2018

### Load libraries 

```python
import pandas as pd # package for high-performance, easy-to-use data structures and data analysis
import numpy as np # fundamental package for scientific computing with Python
import matplotlib
import matplotlib.pyplot as plt # for plotting
import seaborn as sns # for making plots with seaborn



# Print all rows and columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

```

### Read the data

```python
df = pd.read_csv('data/gun-violence-data_01-2013_03-2018.csv', parse_dates=["date"])
```

### Checking for missing data

```python
total_null = df.isnull().sum().sort_values(ascending = False)
percent_null = (df.isnull().sum()/df.isnull().count()*100).sort_values(ascending = False)
null_df_data = pd.concat([total_null, percent_null], axis=1, keys=['Total', 'Percent'])
null_df_data.head(29)
```

# Data Analysis 

### States with highest total of gun violence events (injuries/deaths)

```python
plt.figure(figsize=(20,8))
plt.xticks(rotation=90)
sns.countplot(x='state', data=df, order=df['state'].value_counts().index).set_title('States with hightest total of gun violence events (injuries/deaths)')
```

```python
df['state'].value_counts()
```

### Cities with hightest total of gun violence events (injuries/deaths)

```python
plt.figure(figsize=(20,8))
plt.xticks(rotation=90)
sns.countplot(x='city_or_county', data=df, order=df['city_or_county'].value_counts().head(30).index).set_title('Cities with hightest total of gun violence events (injuries/deaths)')
```

```python
df['city_or_county'].value_counts().head(30)
```

### Total guns involved during violent event (injuries/deaths)

```python
df['n_guns_involved_categorized'] = df['n_guns_involved'].dropna().apply(lambda x: "4+" if x>4 else str(x))

plt.figure(figsize=(20,8))
plt.xticks(rotation=90)
sns.countplot(x='n_guns_involved_categorized', data=df, order=df['n_guns_involved_categorized'].value_counts().index).set_title('Total guns involved during violent event (injuries/deaths)')
```

```python
df['n_guns_involved_categorized'].value_counts()
```

### States with most people killed by guns

```python
df[['state', 'n_killed']].groupby(['state'], 
                                   as_index=False)
```

```python
plt.figure(figsize=(20,8))
plt.xticks(rotation=90)
df_sorted_n_killed_by_state = df[['state', 'n_killed']].groupby(['state'], 
                                   as_index=False).sum().sort_values(by='n_killed', ascending=False).head(20)

sns.barplot(x='state', y='n_killed', data=df_sorted_n_killed_by_state)


```

```python
df[['state', 'n_killed']].groupby(['state'], as_index=False).sum().sort_values(by='n_killed', ascending=False).head(20)
```

### Cities/counties with most people killed by guns

```python
plt.figure(figsize=(20,8))
plt.xticks(rotation=90)
df_sorted_n_killed_by_city_county = df[['city_or_county', 'n_killed']].groupby(['city_or_county'],
                                    as_index=False).sum().sort_values(by='n_killed', ascending=False).head(20)

sns.barplot(x='city_or_county', y='n_killed', data=df_sorted_n_killed_by_city_county)
```

```python
df[['city_or_county', 'n_killed']].groupby(['city_or_county'],as_index=False).sum().sort_values(by='n_killed', ascending=False).head(20)
```
